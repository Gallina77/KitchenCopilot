```python
import pdfplumber as pdf
import re
```

# 1. Load the Sales Report

Load the Sales Report in the folder Verkaufsberichte

```python
with pdf.open("Verkaufsberichte/Taegl_Verkaufsbericht_Restaurant251108143033.pdf") as sales_report:
    for page in sales_report.pages:
        sales_report_text += page.extract_text()
print(sales_report_text)        
        
```

# 2. Identify the Start and End Date of the Sales Report
Detect and store the Start and End Date

```python
date_range_pattern = r"Berichtsbereich\s+von\s+(\d{2}\.\d{2}\.\d{4})\s+bis\s+(\d{2}\.\d{2}\.\d{4})"
match = re.search(date_range_pattern, sales_report_text)
if match:
    start_date = match.group(1)  # '08.11.2025'
    end_date = match.group(2)    # '08.11.2025'
    print(f"Start: {start_date}, End: {end_date}")
```

# 3. Find the Sales Total 

Detect and store the Sales Total of that day

```python
total_sales_pattern = "^Verkauf.*Gesamtergebnis$"
#print(re.search(total_sales_pattern, sales_report_text, re.MULTILINE))
for line_number, line_text in enumerate(sales_report_text.split('\n')):
   # print(line[0], line[1])
    if re.search(total_sales_pattern, line_text, re.MULTILINE):
        matching_line = line_number

sales_numbers = sales_report_text.split('\n')[matching_line+1].split()
sales_total = sales_numbers[0]
print(sales_total)

   
```
